```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
```

```python
df = pd.read_excel('/content/relation_medak.xlsx')
df.columns
```

```python
df['EC_Result']
```

```python
# Assuming df is your dataframe with columns A, P, Q, R, S, T, U, V, W

# Extracting columns A, P, Q, R, S, T, U, V, W
columns_to_plot = ['EC__mS_cm_', 'NDSI', 'SI1', 'SI2', 'SI3', 'SI4', 'SI5', 'SI6']
df_subset = df[columns_to_plot]
```

```python
# Plotting the relationships
fig, axes = plt.subplots(len(columns_to_plot) - 1, 1, figsize=(8, 20))
for i, column in enumerate(columns_to_plot[1:]):  # Excluding column A
    axes[i].scatter(df_subset[column], df_subset['EC__mS_cm_'],  alpha=0.5)
    axes[i].set_ylabel('EC__mS_cm_')
    axes[i].set_xlabel(column)

plt.tight_layout()
plt.show()
```
